# Programa que pergunta valores de x para equação quadratica

```python
import math
import sys

print('Informe o valor (ou valores) de x para a seguinte equação: x² + 2x - 20 = 0: ')
x1User = float(input('x1: '))
x2User = float(input('x2: '))

a=1
b=2
c=-20
d=b**2-(4*a)*c
m1=math.sqrt(d)
x1=(-b+m1)/(2*a)
x2=(-b-m1)/(2*a)
x1 = round(x1, 2)
x2 = round(x2, 2)

if(((x1User == x1)or(x1User == x2))and((x2User == x1)or(x2User == x2))):
    print('Parabéns!')
else:
    print('Essa não é bem a resposta correta, a resposta correta é:', x1, 'e ', x2)
```

# Função para plotar gráfico que mostra a parábola da função quádratica

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

def Graph(x1, fx1):
    lin = x1 + x1/3
    
    x = np.linspace(-lin, lin, 100)
    y = x**2 + 2*x - 20 

    plt.plot(x, y)
    plt.plot(x1, fx1, 'or')

    plt.title('x² + 2x - 20')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.grid()

    plt.show() 
```

# Programa que recebe valor de x e f(x) para a função quadrática

```python
x = float(input('Escolha um valor de x: '))
fx = x**2 + 2*x - 20
fxInput = input('Informe f(x) de x² + 2x - 20 se x = '+ str(x) + ': ')

if(fx == float(fxInput)):
    print('Parabéns!')
else:
    print('Essa não é bem a resposta correta, a resposta correta é:', fx)

Graph(x, fx)
```
